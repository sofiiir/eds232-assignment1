## 5.

Compute summary statistics — mean, median, min, max, and std — for `dna_richness` and `trad_richness`. Store the result in a data frame called `stats` with the following row names `mean`, `median`, `min`, `max`, `std` and column names `dna_richness` and `trad_richness`.



````python
# Calculate all of the stats for dna_richness
dna_richness_mean = df['dna_richness'].mean()
dna_richness_median = df['dna_richness'].median()
dna_richness_min = df['dna_richness'].min()
dna_richness_max = df['dna_richness'].max()
dna_richness_std = df['dna_richness'].std()

# Calculate all of the stats for trad_richness
trad_richness_mean = df['trad_richness'].mean()
trad_richness_median = df['trad_richness'].median()
trad_richness_min = df['trad_richness'].min()
trad_richness_max = df['trad_richness'].max()
trad_richness_std = df['trad_richness'].std()


# Create a dictionary for the dna_richness and trad_richness stats
stats_dict = {
 "dna_richness": [dna_richness_mean, 
                  dna_richness_median,
                  dna_richness_min,
                  dna_richness_max,
                  dna_richness_std], 
  "trad_richness" : [trad_richness_mean, 
                  trad_richness_median,
                  trad_richness_min,
                  trad_richness_max,
                  trad_richness_std]
}

# Create a dataframe from the dictionary 
stats = pd.DataFrame(stats_dict, index = ["mean", "median", "min", "max", "std"])

# View the dataframe
stats
````

## 6.

How many lakes had eDNA detect **more** species than conventional surveys? How many had conventional surveys detect **more** species? How many were **tied**?

Store your answers in variables called `edna_greater`, `trad_greater`, and `tied`.

````python
# Calculate how many surveys have more eDNA detected species than conventionally detected species
edna_greater = len(df[df['dna_richness'] > df['trad_richness']])

# Calculate how many surveys have more conventionally detected species than eDNA detected species
trad_greater = len(df[df['trad_richness'] > df['dna_richness']])

# Calculate how many surveys have the same number of conventionally detected species and eDNA detected species
tied         = len(df[df['dna_richness'] ==  df['trad_richness']])
````

## 10.

Group lakes by `gear_cat` (single vs. multiple conventional survey methods) **and** `marker_cat` (single vs. multiple eDNA markers). Using `gear_cat` as rows and `marker_cat` as columns, compute a pivot table showing the mean `union` (total species richness) and mean `dna_advantage` for each combination. Round values to 2 decimal places and exclude the `unspecified` category from your interpretation.

Store the result in a variable called `pivot`.

**Hint — `pivot_table` syntax:**
```python
df.pivot_table(
    values=[...],      # columns to aggregate
```
